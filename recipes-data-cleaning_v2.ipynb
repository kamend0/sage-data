<h1>Data Cleaning</h1>

```python
import pandas as pd
import ast

print("Libraries imported successfully.")
```

```python
recipes = pd.read_csv("./raw-data/recipes-data.csv",
                     index_col = 0)
```

```python
recipes.head()
```

```python
# rec = ast.literal_eval(recipes['Cleaned_Ingredients'][0])

clean_recipes = recipes.copy()

# Replace Ingredients column with Cleaned Ingredients column
clean_recipes.drop(columns = ["Ingredients"], 
                   inplace = True)
clean_recipes.rename(columns = {"Cleaned_Ingredients":"Ingredients"},
                    inplace = True)

# Rearrange
clean_recipes = clean_recipes[["Title", "Ingredients", "Instructions", "Image_Name"]]

clean_recipes.head()
```

```python
# Ingredients column is actually just full of strings.
# Need to convert to be actual lists.

clean_recipes['Ingredients'] = clean_recipes['Ingredients'].apply(ast.literal_eval)
clean_recipes.head()
```

```python
print(type(clean_recipes['Title'][0])) # Should be str
print(type(clean_recipes['Ingredients'][0])) # Should be list
print(type(clean_recipes['Instructions'][0])) # Should be str
print(type(clean_recipes['Image_Name'][0])) # Should be str
```

```python
# Pan-seared salmon on baby arugula needs to be removed here,
# because the image file is corrupted, and I just don't wanna
# mess with that.

clean_recipes.drop(clean_recipes[clean_recipes['Title'] == "Pan Seared Salmon on Baby Arugula"].index,
                   inplace = True)
clean_recipes.reset_index(drop = True, inplace = True)
# clean_recipes.drop(columns=['index'], inplace = True)
```

```python
# Check to make sure recipe was deleted

clean_recipes[9890:9900]
```

```python
# Rename columns just to clean everything up
clean_recipes.rename(columns = { 'Title' : 'title',
                                 'Ingredients' : 'ingredients',
                                 'Instructions' : 'instructions',
                                 'Image_Name' : 'imageName' },
                     inplace = True)

clean_recipes.head()
```
